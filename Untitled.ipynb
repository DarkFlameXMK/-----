```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate
a = np.array([0,0.3,0.6,0.9,1.2,1.5])
b = np.array([0.61,0.55,0.55,0.55,0.55,0.55])
c = np.array([0.61,0.56,0.55,0.55,0.55,0.54])
d = np.array([0.61,0.59,0.55,0.55,0.55,0.55])
a_new = np.linspace(0,1.5,300)
tckb = interpolate.splrep(a, b)
b_bspline = interpolate.splev(a_new, tckb)
tckc = interpolate.splrep(a, c)
c_bspline = interpolate.splev(a_new, tckc)
tckd = interpolate.splrep(a, d)
d_bspline = interpolate.splev(a_new, tckd)
```

```python
from matplotlib import pyplot as plt
import numpy as np
import matplotlib
fig = plt.figure(figsize=(10,6),dpi=120)
fig1 = fig.add_subplot(2,2,1)


plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False   

fig1.plot(a_new,b_bspline,label='内层当量比1.0',color = 'blue',linestyle = '--',alpha=.6)
fig1.plot(a_new,c_bspline,label='内层当量比0.8',color = 'red',linestyle = ':',alpha=.6)
fig1.plot(a_new,d_bspline,label='内层当量比0.6',color = 'green',alpha=.6)
plt.ylim((0.5,0.7))
plt.xlabel('内层流量L/min',fontproperties="SimHei")
plt.ylabel('外层可燃极限φ',fontproperties="SimHei")
plt.legend(loc='upper right')
plt.savefig('./t1.svg')
plt.show()
```
